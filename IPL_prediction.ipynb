Encoding the categorial data

```python
ipl_df['city'] = ipl_df['city'].astype('category').cat.codes
ipl_df['team1'] = ipl_df['team1'].astype('category').cat.codes
ipl_df['team2'] = ipl_df['team2'].astype('category').cat.codes
ipl_df['venue'] = ipl_df['venue'].astype('category').cat.codes
ipl_df['toss_winner'] = ipl_df['toss_winner'].astype('category').cat.codes
ipl_df['toss_decision'] = ipl_df['toss_decision'].astype('category').cat.codes
ipl_df['result'] = ipl_df['result'].astype('category').cat.codes
ipl_df['player_of_match'] = ipl_df['player_of_match'].astype('category').cat.codes
ipl_df['venue'] = ipl_df['venue'].astype('category').cat.codes

ipl_df['winner'] = ipl_df['winner'].astype('category').cat.codes
```

```python
ipl_df.drop('umpire1', axis=1, inplace=True)
ipl_df.drop('umpire2', axis=1, inplace=True)
ipl_df.drop('umpire3', axis=1, inplace=True)
```

```python
ipl_df['date'] = pd.to_datetime(ipl_df['date'])
ipl_df.head()
```

```python
ipl_df['Day'] = ipl_df['date'].dt.day
ipl_df['Month'] = ipl_df['date'].dt.month
ipl_df.head()
```

```python
ipl_df.drop('date', axis=1, inplace=True)
ipl_df.head()
```

Separating or splitting features and the target

```python
X = ipl_df.drop(['winner'], axis=1)
Y = ipl_df['winner']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=8)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

```python
# Initialize the Random Forest model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
rf_model.fit(X_train, y_train)
```

Model Evaluation

```python
# Prediction on training data
y_pred = rf_model.predict(X_train)
```

```python
accuracy = accuracy_score(y_train, y_pred)
print("Accuracy:", accuracy)

print("Classification Report:\n",classification_report(y_train, y_pred))
```

```python
# Prediction on testing data
y_pred = rf_model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

print("Classification Report:\n",classification_report(y_test, y_pred))
```
